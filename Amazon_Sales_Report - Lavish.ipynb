```python
#drop null columns
data.drop(["PendingS","New"], axis = 1, inplace = True)
```

```python
#Drop all Values
data.dropna(inplace= True)
```

```python
# change data type
data["ship-postal-code"] = data["ship-postal-code"].astype("int")
```

```python
#Convert date to datetime object
data["Date"] = pd.to_datetime(data["Date"])
```

```python
#Rename is the column name
data.rename(columns={'Qty':'Quantity'}, inplace = True)
```

# Exploratory Data Analysis

## Size

```python
data["Size"].value_counts()
```

```python
#ploting a countplot for size
ax = sns.countplot(x = "Size", data = data) 
for bars in ax.containers:
    ax.bar_label(bars)
```

##### From the above Graph we can see that Most of peoples are buying M Size

```python
# Grouping the data by "Size", summing the "Quantity" for each group and sorting in descending order

S_Qty = data.groupby(["Size"], as_index = False) ["Quantity"].sum().sort_values("Quantity", ascending = False)
```

```python
S_Qty
```

```python
#ploting a bar chart for Size and Quantitty

sns.barplot(x = "Size", y = "Quantity", data = S_Qty)
```

##### From above Graph we can say that most of buy Quantity is M Size in sales

## Status

```python
status_count = data["Status"].value_counts(ascending = False)
status_count
```

```python
#ploting a countplot for status to visualize

plt.figure(figsize = (8,6))   
sns.countplot(y = "Status", data = data , palette = "Paired")
```

```python
status_count = data["Status"].value_counts().head(4)
status_count
```

```python
#ploting a pie chart to show top 4 Status
plt.pie(x= status_count, labels = status_count.index,autopct='%1.1f%%' , startangle=70)
plt.show()
```

##### From above chart we can say that 78.1% shipped-Delivered to Buyer and 13.9% is cancelled

## Courier Status

```python
#ploting a countplot for courier status to visualize

plt.figure(figsize=(11,6))
sns.countplot(x= "Courier Status", data = data , hue = "Status")
```

#####  From above Graph the majority of the orders are shipped through the courier.

```python
Courier_count =data["Courier Status"].value_counts()
Courier_count
```

```python
#ploting a pie chart to show courier status
plt.pie(x= Courier_count, labels = Courier_count.index, autopct='%1.1f%%', startangle=90)
plt.show()
```

## Category

```python
data["Category"].dtype
```

```python
#ploting a countplot for Category to visualize
plt.figure(figsize = (9,6))
sns.countplot(x = "Category", data = data)
```

##### From above Graph we can see that most the buying Category is T-Shirt

```python
# Grouping the data by "Category", summing the "Amount" for each group and sorting in descending order

Category_sales = data.groupby(["Category"], as_index = False) ["Amount"].sum().sort_values("Amount", ascending = False)
```

```python
#ploting a bar chart for Category and Amount

plt.figure(figsize = (16,7))
ax = sns.barplot(x = "Category", y = "Amount", data = Category_sales)
for bars in ax.containers:
    ax.bar_label(bars)
```

##### From above chart we can say that Top selling category is "T-Shirt", "Shirt and ""Blazzer"

## ship-state

```python
#ploting a countplot for ship-state

plt.figure(figsize=(12,6))
sns.countplot(x= "ship-state", data= data)
plt.xticks(rotation =90)
plt.show()
```

```python
#Top_10_State
Top_10_State = data["ship-state"].value_counts().head(10)
Top_10_State
```

```python
#ploting a barchart for Top 10 states
Top_10_State.plot(kind="bar")
plt.title("Distribution of State")
plt.ylabel("Number")
plt.xlabel("States")
```

```python
# Grouping the data by "ship-state", summing the "Amount" for each group and sorting in descending order

ship_state = data.groupby(["ship-state"], as_index = False) ["Amount"].sum().sort_values("Amount", ascending = False).head(10)
ship_state
```

```python
#ploting a bar chart for ship-state and Amount
plt.figure(figsize=(16,6))
sns.barplot(x= "ship-state", y = "Amount", data = ship_state, palette = "bright")
```

##### From above graph we can see that High Sales Value Comes form Maharashtra State
